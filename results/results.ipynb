```python
import numpy as np
import matplotlib.pyplot as plt

font = {
    "size": 20
}
```

# CALCE数据集

```python
calce_dataset_path = "../data/processed_data/calce/calce_capacity.npy"
calce_gt = np.load(calce_dataset_path, allow_pickle=True)
calce_battery_names = ["CS2_35", "CS2_36", "CS2_37", "CS2_38"]
b35_gt = calce_gt[0]
b36_gt = calce_gt[1]
b37_gt = calce_gt[2]
b38_gt = calce_gt[3]
```

```python
# AR
b35_ar = np.load("./calce_CS2_35_ar.npy", allow_pickle=True)
b36_ar = np.load("./calce_CS2_36_ar.npy", allow_pickle=True)
b37_ar = np.load("./calce_CS2_37_ar.npy", allow_pickle=True)
b38_ar = np.load("./calce_CS2_38_ar.npy", allow_pickle=True)
calce_ar_pred = (b35_ar, b36_ar, b37_ar, b38_ar)
```

```python
# SVR
b35_svr = np.load("./calce_CS2_35_svr.npy", allow_pickle=True)
b36_svr = np.load("./calce_CS2_36_svr.npy", allow_pickle=True)
b37_svr = np.load("./calce_CS2_37_svr.npy", allow_pickle=True)
b38_svr = np.load("./calce_CS2_38_svr.npy", allow_pickle=True)
calce_svr_pred = (b35_svr, b36_svr, b37_svr, b38_svr)
```

```python
# MLP
b35_mlp = np.load("./calce_CS2_35_mlp.npy", allow_pickle=True)
b36_mlp = np.load("./calce_CS2_36_mlp.npy", allow_pickle=True)
b37_mlp = np.load("./calce_CS2_37_mlp.npy", allow_pickle=True)
b38_mlp = np.load("./calce_CS2_38_mlp.npy", allow_pickle=True)
calce_mlp_pred = (b35_mlp, b36_mlp, b37_mlp, b38_mlp)
```

```python
# LSTM
b35_lstm = np.load("./calce_CS2_35_lstm.npy", allow_pickle=True)
b36_lstm = np.load("./calce_CS2_36_lstm.npy", allow_pickle=True)
b37_lstm = np.load("./calce_CS2_37_lstm.npy", allow_pickle=True)
b38_lstm = np.load("./calce_CS2_38_lstm.npy", allow_pickle=True)
calce_lstm_pred = (b35_lstm, b36_lstm, b37_lstm, b38_lstm)
```

```python
# CNN
b35_cnn = np.load("./calce_CS2_35_cnn.npy", allow_pickle=True)
b36_cnn = np.load("./calce_CS2_36_cnn.npy", allow_pickle=True)
b37_cnn = np.load("./calce_CS2_37_cnn.npy", allow_pickle=True)
b38_cnn = np.load("./calce_CS2_38_cnn.npy", allow_pickle=True)
calce_cnn_pred = (b35_cnn, b36_cnn, b37_cnn, b38_cnn)
```

```python
for idx, gt in enumerate((b35_gt, b36_gt, b37_gt, b38_gt)):
    plt.figure(figsize=(8, 6))
    plt.plot(gt)
    plt.plot(calce_ar_pred[idx])
    plt.plot(calce_svr_pred[idx])
    plt.plot(calce_mlp_pred[idx])
    plt.plot(calce_lstm_pred[idx])
    plt.plot(calce_cnn_pred[idx])
    plt.xlabel("循环圈数", font)
    plt.ylabel("放电容量（Ah）", font)
    plt.legend([f"{calce_battery_names[idx]}真值", "AR模型估计结果", "SVR模型估计结果", "MLP模型估计结果", "LSTM模型估计结果", "CNN模型估计结果"])
    plt.savefig(f"calce_{calce_battery_names[idx]}.jpg", dpi=1000, bbox_inches="tight")
```
